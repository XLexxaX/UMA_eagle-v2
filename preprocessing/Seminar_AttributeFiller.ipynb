# Lowercase everything

```python
import re
import os

f = open(PATH_TO_FILE.replace(".nt","")+"_filled2.nt", 'w', encoding="UTF-8")
with open(PATH_TO_FILE.replace(".nt","")+"_filled.nt", "r", encoding="UTF-8") as file:
    for line in file:
        try:
            lowered_prop = str(line.split('>')[2].split("\"")[1]).lower()
            line = line.split('>')
            expression = line[1].split('http://www.w3.org/1999/02/22-rdf-syntax-ns#')[1]
            line[1] = line[1].replace(expression, re.sub('[^a-zA-Z]', '', expression))
            result = line[0] + ">" + line[1] + "> \"" + lowered_prop + "\" .\n"
            f.write(result)
        except:
            f.write(line)

f.flush()
f.close()
print("Done")
```
